### Práctica 2. Generación de números aleatorios por el método congruencial lineal

```python
# Parámetros del método congruencial lineal
x0 = 7    # Semilla inicial
a = 5     # Multiplicador
c = 3     # Incremento
m = 16    # Módulo
n = 100   # Cantidad de números a generar

def metodo_congruencial_lineal(x0, a, c, m, n): # Definición de la función con parámetros
    numeros = []
    x = x0
    
    for _ in range(n):
        x = (a * x + c) % m
        numeros.append(x / m)  

    return numeros

numeros_generados = metodo_congruencial_lineal(x0, a, c, m, n)

# Imprimir los números generados
for num in numeros_generados:
    print("Número aleatorio: " , num)
```
