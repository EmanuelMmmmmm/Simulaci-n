## Problema 2: Distribución Exponencial 

Contexto: Una estación de carga rápida para vehículos eléctricos quiere simular el tiempo entre la llegada de los vehículos para prever la cantidad de puntos de carga necesarios durante el día. La llegada de los vechículos sigue una distribución exponencial con una tasa de 1 llegada cada 15 minutos, en promedio.
Objetivo: Generar tiempos entre llegadas de vehículos eléctricos durante un día de operación (12 horas)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial (tasa de llegadas)
lambda_llegadas = 1 / 15  # En promedio, 1 llegada cada 15 minutos

# Duración del día de operación en minutos
tiempo_operacion = 12 * 60  # 12 horas en minutos

# Generar tiempos entre llegadas
num_llegadas = 1000  # Número máximo de llegadas simuladas
tiempo_actual = 0
llegadas = []

while tiempo_actual < tiempo_operacion:
    u = np.random.uniform(0, 1)  # Generar número aleatorio uniforme
    tiempo_entre_llegadas = -1 / lambda_llegadas * np.log(1 - u)  # Aplicar transformada inversa
    tiempo_actual += tiempo_entre_llegadas
    if tiempo_actual < tiempo_operacion:
        llegadas.append(tiempo_entre_llegadas)

# Mostrar los tiempos entre llegadas
print("Tiempos entre llegadas generados:", llegadas)

# Graficar los tiempos entre llegadas simulados
plt.hist(llegadas, bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos entre Llegadas Simulados (Exponencial)')
plt.xlabel('Tiempo entre llegadas (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```
